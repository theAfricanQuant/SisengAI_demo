```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate Dates (100 consecutive days)
start_date = "2024-01-01"
num_days = 100
dates = pd.date_range(start=start_date, periods=num_days, freq="D")

# Step 2: Generate Random Data using NumPy RNG
rng = np.random.default_rng(42)  # Seed for reproducibility
random_values = rng.normal(loc=0, scale=1, size=num_days)  # Normal distribution (mean=0, std=1)

# Step 3: Create Pandas DataFrame
df = pd.DataFrame({"Date": dates, "Value": random_values})

# Step 4: Plot the Data
plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Value"], label="Random Values", marker="o", linestyle="-")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Random Time Series Data")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python


r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rticks([0.5, 1, 1.5, 2])
ax.grid(True)

plt.show()
```
